```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# 1. 定义和原脚本一致的 transform
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# 2. 加载数据
dataset = datasets.ImageFolder(root='data/UCMerced_LandUse/Images', transform=transform)
print(f"类别数: {len(dataset.classes)}")
print(f"类别名称: {dataset.classes[:5]} ...")

# 3. 获取一批数据
loader = torch.utils.data.DataLoader(dataset, batch_size=8, shuffle=True)
images, labels = next(iter(loader))
print("图像张量形状:", images.shape)
print("标签形状:", labels.shape)

# 4. 可视化
# 将 [-1,1] 映射回 [0,1] 以便显示
img = images[0] * 0.5 + 0.5
plt.imshow(img.permute(1, 2, 0))
plt.title(dataset.classes[labels[0]])
plt.axis('off')
plt.show()
```
